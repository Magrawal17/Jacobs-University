```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

mean = np.array([[0], [1]])
sigma = np.array([[1, 0.25], [0.5 , 1]])


def gaussian(X,Y):
    z_values = []
    for i in range(0,100):
        newlayer = []
        for x,y in zip(X[i], Y[i]):
            difference = np.array([[x - mean[0][0]], [y - mean[1][0]]])
            coffiecient = (1 / 2 * np.pi * (np.linalg.det(sigma))**(1/2))
            sigma_inverse = np.linalg.inv(sigma)
            index = (-0.5) * np.dot(difference.transpose(), np.dot(sigma_inverse, difference))
            newlayer.append(coffiecient * np.exp(index))
        z_values.append(newlayer)
    return np.array(z_values)

f = lambda x, y: x ** 2 + y ** 2
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x,y)
Z = gaussian(X, Y)[:,:,0,0]
fig = plt.figure(figsize = (15,10))
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, Z,cmap=cm.coolwarm,linewidth=0)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
plt.show()
```
