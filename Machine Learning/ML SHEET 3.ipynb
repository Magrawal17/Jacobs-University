```python

def uninterrupted_capital_chars(email:str):
    pos = 0
    sequences = []
    while pos < len(email):
        start, end = pos, pos
        word = ""
        while email[pos].isupper():
            word += email[pos]
            pos += 1
        if len(word):
            sequences.append(word)
        pos += 1
    return sequences
```

```python
def feature_extract(email:str):
    feature_vector = []
    capitalized = email.upper()
    for word in spam_words:
        frequency = capitalized.count(word.upper()) / (capitalized.count(' ') - 1) 
        feature_vector.append(frequency * 100)
    non_white_space_chars = len(capitalized) - capitalized.count(' ')
    for letter in least_used_letters:
        frequency = (capitalized.count(letter)/non_white_space_chars) * 100
        feature_vector.append(frequency)
    capital_seq = uninterrupted_capital_chars(email)
    avg = len(reduce(lambda a, b: a + b, capital_seq)) / len(capital_seq)
    feature_vector.append(avg)
    max_len = max(list(map(lambda a: len(a), capital_seq)))
    feature_vector.append(max_len)
    capital_letters = len(list(filter(lambda a: a.isupper(), list(email))))
    feature_vector.append(capital_letters)
    return feature_vector
```
